We can use histogram to visualize categorical variable. In this data set we can visualize vehicles with automatic vs manual transimission. We will use the above functions. plt.ylim function- to define the range of the y axis

We can calculate max, min, count, len and mean of disp variable

```python
import numpy as np
mtcarsDf['disp'].apply({'min':min,'max':max,'count':len,'mean':np.mean})
```

We can calculate the sum, mean,min and max, and len of the "hp" variable in the data set

```python
mtcarsDf['hp'].aggregate({'sum':sum,'min':min,'max':max,'count':len,'mean':np.mean})
```

We can calculate min, max, mean, count of all the variable in the data set using aggregate.

```python
mtcarsDf.aggregate(['sum','min', 'max', 'mean', 'count'])
```

To transform data using the Pandas library, you can use the DataFrame.apply() method to apply a function to each row or column of a DataFrame. You can use this method to transform the values in a column or to create new columns based on the values in existing columns.

For example, to create a new column in the mtcars data set that contains the weight of each car in pounds, you could use the following code:

```python
# Create a new column that contains the weight of each car in pounds
# 1 kilogram is equal to 2.205 pounds
mtcarsDataFrame["wt_lbs"] = mtcarsDataFrame["wt"].apply(lambda x: x * 2.205)

# Print the first 5 rows of the updated data set
print(mtcarsDataFrame.head())
```
